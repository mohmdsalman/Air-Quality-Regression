# after outlier remove

```python
df1=data
```

```python
df1.drop(columns=['City', 'NO', 'NO2', 'NH3', 'O3',
       'Benzene', 'Toluene', 'Year', 'Month', 'Day_Of_Week'], inplace=True)
```

```python
df1
```

```python
df1['AQI'] = np.log1p(df1['AQI'])
```

```python
def remove_outliers_iqr(df1, exclud_column):
    for col in df1.columns:
      if col == exclud_column:
          continue
      Q1 = df1[col].quantile(0.25)
      Q3 = df1[col].quantile(0.75)
      IQR = Q3 - Q1
      lower_bound = Q1 - 1.5 * IQR
      upper_bound = Q3 + 1.5 * IQR
      # Filter the DataFrame to remove rows with outliers in each column
      df1 = df1[(df1[col] >= lower_bound) & (df1[col] <= upper_bound)]
    return df1

# Specify the column to exclude
df1_cleaned = remove_outliers_iqr(df1, exclud_column="AQI")
df1_cleaned
```

```python
x = df1_cleaned[['PM2.5', "PM10", 'NOx', 'CO', 'SO2']]
y = df1_cleaned["AQI"]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)


print(f"MSE: {mse}, RMSE: {rmse}, R-squared: {r2}, MAE: {mae}")
```
